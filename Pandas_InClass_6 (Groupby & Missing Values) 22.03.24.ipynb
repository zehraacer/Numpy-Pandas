**None**:

   -It means "nothing" in Python.

   -It is used to indicate that an object has no value in Python.

   -Its type is NoneType and it has only a single value: None.

   -For example, if a function does not return any value, Python automatically returns None.

   
**nan ve NaN**:

   - It stands for "Not a Number."

   -It represents a value of the float type.

   -It is used to indicate non-numeric values.

   -In libraries such as Pandas and Numpy, it is also used to represent missing data.

   -When comparing nan values with equality, it always returns False. That is, the expression nan == nan results in False.

   -There is no technical difference between nan and NaN. Different sources or libraries may prefer different spellings, but they mean the same thing.

**summary**:

   -None is a unique value in Python and means "no value."
   
   -nan or NaN represents non-numeric values and is often used for missing data or mathematically undefined operations.

These two values belong to different types and are used for different situations.

In practice, you often encounter NaN values in Pandas when performing data analysis. However, None is more common in general Python programming.

```python
id_no = ["P001", "P002", "P003", "P004", "P005", "P006", "P007", "P008", "P009", "P010", "P011"]
gender = ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F", "M"]
status = ["FT", "PT", "-", "FT", "PT", "PT", "FT", "-", "PT", "FT", np.nan]
dept = ["DS", "FS", "AWS", "AWS", "DS", None, "FS", "FS", np.nan, "DS", "AWS"]
V1 = np.array([2, 3, 5, np.nan, 7, 1, np.nan, 10, 14, "-", 6])
V2 = np.array([8, np.nan, 5, 8, 11, np.nan, np.nan, 2, 3, 7, 9])
salary = np.array([np.nan, 54, 59, 120, 58, 75, None, 136, 60, 125, np.nan])

df0 = pd.DataFrame({
                   "id" : id_no,
                   "gender": gender,
                   "status": status,
                   "dept": dept,
                   "var1" : V1,
                   "var2" : V2,
                   "salary" : salary
                  })
```

```python
print("1", 0 * np.nan)
print("2", np.nan - np.nan)
print("3", np.nan + np.nan)
print("4", np.nan - 10)
print("5", np.nan + 10)
print("6", np.nan == np.nan) # not equal in value, because it's undefined.
print("7", 10 > np.nan)
print("8", np.inf > np.nan) # even infinity > nan returns false because the value is undefined
print("9", float('-inf') < np.nan < float('inf')) # nan is also not a value that falls between negative and positive infinity.
print("10", np.nan is np.nan) # when queried with "is", it checks the id values, i.e.;
# memory location, which are the same.
# However, this does not mean the values are the same.
```

## <font color='blue'> <b>dropna() & drop()</b><font color='black'>

```python
df
```

```python
df.dropna() #It deletes rows that contain nan values.
```

```python
df.dropna(axis = 0, how= "any", inplace= False) #An "any" axis method will delete any rows with at least one NaN value, and if inplace=True, the change will be permanent.
```

```python
df.dropna(axis = 0, how= "all", inplace= False) #The "all" axis method will delete rows where all values are NaN, and with inplace=True, the change will be permanent.
```

```python
df.dropna(axis = 1, how= "any", inplace= False)
```

```python
df.dropna(axis = 1, thresh= 9, inplace= False) #I want it to return columns with at least 9 non-NaN values. I specify the number of values with 'thresh'.
```
