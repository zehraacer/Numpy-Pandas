## <font color='blue'> <b>reset_index() & set_index()</b><font color='black'>

set_index() sets the specified column or columns as the new index of the DataFrame.

# <font color='green'> <b>Groupby & Aggregation Methods (aggregate() & agg())</b><font color='black'>
    
groupby(); is used to group data based on a specific column in a dataset.

This function is used to divide data into groups and perform various operations on these groups.

For example, in a dataset containing customer names, ages, and purchased products, you can use the groupby() function to group customers based on the products they purchased and find the average age for each product.

[Source01](https://towardsdatascience.com/all-pandas-groupby-you-should-know-for-grouping-data-and-performing-operations-2a8ec1327b5),
[Source02](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html),
[Source03](https://naomi-fridman.medium.com/pandas-groupby-explained-with-titanic-6a1c47eb8182)
Pandas' aggregation methods perform operations on a specific dataset to return a result. These methods allow for various operations such as summing values, calculating averages, finding maximum or minimum values, among others.

Some of Pandas' aggregation methods include:

**sum():** Returns the sum of all values in the specified column.

**mean():** Returns the mean (average) of all values in the specified column.

**median():** Calculates the median of the grouped data.

**mode():** Returns the most frequently occurring value(s) in the specified column.

**max():** Returns the maximum value in the specified column.

**min():** Returns the minimum value in the specified column.

**count():** Returns the total number of values in the specified column.

**std():** Returns the standard deviation of the values in the specified column.

**var():** Returns the variance of the values in the specified column.

**describe():** Computes a statistical summary of the grouped data.

**idxmin():** Returns the index of the first occurrence of the minimum value along the specified axis.

**idxmax():** Returns the index of the first occurrence of the maximum value along the specified axis.

**corr():** Computes the correlation between columns, excluding NA/null values.

These methods can be applied on a DataFrame or Series, allowing operations on desired columns or rows.

Pandas' aggregation methods are highly useful for efficiently analyzing data.

[Pandas Official Documentation](https://pandas.pydata.org/docs/reference/frame.html)    







It's used to apply different functions (e.g., mean, sum, max) on a dataset grouped by specific column(s) or axis.

[aggregate](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.aggregate.html),
[agg](https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.core.groupby.DataFrameGroupBy.agg.html),

```python
titanic.groupby("sex")["age"].mean()  #Gender-wise average age.
```

```python
titanic.head()
```

```python
titanic.groupby("pclass")["fare"].mean()  #The average fares paid by classes.
```

```python
titanic.groupby("sex")["age"].min()  #It grouped by gender and displayed the least.
```

```python
titanic.groupby("sex")["age"].max()  #It grouped by gender and displayed the most.
```

```python
titanic.groupby("sex")["age"].agg([min, max])  #it provided output as a DataFrame for max and min. We can perform multiple operations with agg().
```

```python
titanic.groupby("sex")["age"].agg([min, max, np.mean])
```

```python
titanic.groupby("sex")["survived"].count()  #We used count to check the number of occurrences.
```

```python
titanic.groupby(["sex", "survived"]).count()
```
